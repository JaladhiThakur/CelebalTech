```
#importing necessary libraries
import datetime
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```
#loading dataset
data=pd.read_excel("/content/drive/MyDrive/Celebal Tech/Assignment 2/premiums.xlsx")
data.head()
```

```
#printing dataset
data.head()
```

```
#sorting based on customer and grouping according to product. The result will give dataset in sorted form
data = data.sort_values(by='Customer')
data = data.groupby(['Product']).apply(lambda x: x.sort_values('From'))
data = data.apply(lambda x: x)
data = data.reset_index(drop = True)
```

```
data.head(200)
```

```
#dropping unnamed index
data = data.drop('Unnamed: 0',axis=1)
data.head()
```

```
#Case 1: Different premium for same customer and same product having same date
for i in range(len(data)):
    for j in range(i,len(data)):
        if((data['Customer'][i] == data['Customer'][j]) and (data['Product'][i] == data['Product'][j]) and (data['Premium'][i] != data['Premium'][j]) and (data['To'][i] != data['To'][j])) and (data['From'][i]< data['From'][j]):
                              data.loc[i,'To'] = data['From'][j]

#case 2: Overlapping problem
disp =[]
for i in range(len(data)):
    for j in range(i,len(data)-i):
        if(((str(data['From'][i])) == str(data['From'][j])) and (str(data['To'][i]) == str(data['To'][j])) and (data['Customer'][i] == data['Customer'][j]) and ((data['Premium'][i] < data['Premium'][j])) and (data['Product'][i] == data['Product'][j])):
            disp.append(i)


data['From'] = pd.to_datetime(data['From'])
data['To'] = pd.to_datetime(data['To'])

#calculating difference between months
diff = data['To'].dt.to_period('M').astype(int) - data['From'].dt.to_period('M').astype(int)

#extracting year and month from full date
data['From'] = pd.to_datetime(data['From']).dt.strftime('%Y-%m')
data['To'] = pd.to_datetime(data['To']).dt.strftime('%Y-%m')

#calculating difference between the months
#def diff_month(d1, d2):
   # return (d1.year - d2.year) * 12 + d1.month - d2.month

#creating output dataframe
output_df = pd.DataFrame(columns=['Customer','Material','Month-Year','Premium'])


#building the output dataframe as per the requirement
for i in range(len(data)):
    
    customer = data['Customer'][i]
    material = data['Product'][i]
    premium = data['Premium'][i]
    start_date = data['From'][i]
    to = data['To'][i]
    dif = diff[i]
    
    #while((diff_month(data['From'][i], data['To'][i])>0)):
    while(dif>0):
      output_df.loc[i,'Customer'] = customer
      output_df.loc[i,'Material'] = material
      output_df.loc[i,'Month-Year'] = start_date
      output_df.loc[i,'Premium'] = premium
      a = start_date
      a = a.split('-')
      year = a[0]
      month = a[1]

      if(int(a[1])+1 > 12):
          year = int(a[0])+1
          month = 1
      else:
          year = int(a[0])
          month = int(a[1])+1

      if(month>=10):
          start_date = str(year) +'-'+str(month)
      else:
          start_date = str(year) +'-'+'0'+str(month)

          dif=dif-1
            
            




```
